```python
#covertir la colonne Date_mutation de type objet en type date
df['Date_mutation'] = pd.to_datetime(df['Date_mutation'], format='%d/%m/%Y')
df.info()
```

```python
#convertir la colonne  Nature_mutation de type objet en type string:
df['Nature_mutation'] = df['Nature_mutation'].apply(str)
df.info()
```

```python
#Convertir la colonne Valeur_fonciere de type objet en type float:
import numpy as np

def nettoyer_valeur(val):
    if isinstance(val, str):  # Vérifier que c'est bien une chaîne
        val = val.replace(' ', '')  # Supprimer les espaces
        val = val.replace(',', '.')  # Remplacer les virgules par des points si besoin
        if 'M' in val:  # Si la valeur est en millions (ex: "1.5M")
            val = val.replace('M', '')
            return float(val) * 1_000_000  # Convertir en valeur numérique
        try:
            return float(val)  # Convertir en float
        except ValueError:
            return np.nan  # Si erreur, mettre NaN
    return np.nan  # Si valeur vide ou autre type, mettre NaN

# Appliquer la fonction à la colonne
df['Valeur_fonciere'] = df['Valeur_fonciere'].apply(nettoyer_valeur)

df.info()
```

```python
#Remplacer la colonne de valeur fonciere nan en type float:
df['Valeur_fonciere'] = df['Valeur_fonciere'].fillna(0.0)
df.head()
```
